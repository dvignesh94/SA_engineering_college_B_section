```python
## Encodes into 1's and 0's
import pandas as pd

# Read the Excel file
data = pd.read_excel(
    "/Users/vignesh/Documents/GitHub/SA_engineering_college_B_section/Data Analysis/corr_analysis_v2.xlsx",
    engine='openpyxl'
)

# Specify categorical variables
categorical_vars = [
    'What grade level is your child in?',
    'What would you like help with today?',
    'What type of support are you interested in?',
    'How much help do you think your child needs to achieve their goal?',
    'How soon are you looking to get started?',
    'Lead Stage'
]

# One-hot encode categorical columns
one_hot_encoded = pd.get_dummies(data[categorical_vars], drop_first=False)

# Concatenate and drop originals
data_encoded = pd.concat([data.drop(columns=categorical_vars), one_hot_encoded], axis=1)

# Convert bool columns to numeric (True→1, False→0)
data_encoded = data_encoded.assign(**{
    col: data_encoded[col].astype(int)
    for col in data_encoded.select_dtypes('bool').columns
})

# Compute correlation matrix
correlation_matrix = data_encoded.select_dtypes(include=['number']).corr()

# Export to CSV
correlation_matrix.to_csv("correlation_matrix.csv", index=True)

print("✅ Correlation matrix generated successfully!")
print("Shape:", correlation_matrix.shape)
print(correlation_matrix.head())
```
